```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.applications import EfficientNetB3
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau

# Mount Google Drive if not already mounted
from google.colab import drive
drive.mount('/content/drive')

# Path to the knee implant folder
base_path = "/content/drive/MyDrive/Knee"

# List all implant types from folder names
implant_types = [f for f in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, f))]
print(f"Found {len(implant_types)} implant types:")
for i, implant in enumerate(implant_types):
    print(f"{i+1}. {implant}")
```

```python
# Count images in each folder
image_counts = {}
total_images = 0

for implant in implant_types:
    folder_path = os.path.join(base_path, implant)
    image_files = [f for f in os.listdir(folder_path)
                  if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    image_counts[implant] = len(image_files)
    total_images += len(image_files)

# Display distribution
print(f"Total images: {total_images}")
for implant, count in image_counts.items():
    print(f"{implant}: {count} images ({count/total_images*100:.1f}%)")

# Check for class imbalance
plt.figure(figsize=(15, 6))
plt.bar(image_counts.keys(), image_counts.values())
plt.xticks(rotation=90)
plt.title("Number of Images per Implant Type")
plt.tight_layout()
plt.show()
```

```python
# Split ratios
train_ratio = 0.7
validation_ratio = 0.15
test_ratio = 0.15

# Set up image generators with augmentation for training
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=validation_ratio + test_ratio
)

# Set up validation and test generators (just rescaling)
valid_test_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=validation_ratio/(validation_ratio + test_ratio)
)

# Image size - using 224x224 which is standard for many pre-trained models
img_height, img_width = 224, 224
batch_size = 16

# Create the generators
train_generator = train_datagen.flow_from_directory(
    base_path,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training',
    shuffle=True
)

validation_generator = valid_test_datagen.flow_from_directory(
    base_path,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training',
    shuffle=False
)

test_generator = valid_test_datagen.flow_from_directory(
    base_path,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation',
    shuffle=False
)

# Get class indices mapping
class_indices = train_generator.class_indices
idx_to_class = {v: k for k, v in class_indices.items()}

# Save class indices for later use
import json
with open('knee_implant_classes.json', 'w') as f:
    json.dump(idx_to_class, f)
```

```python
# Create a base model from EfficientNetB3
base_model = EfficientNetB3(
    weights='imagenet',
    include_top=False,
    input_shape=(img_height, img_width, 3)
)

# Freeze the base model layers
base_model.trainable = False

# Create the model architecture
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(512, activation='relu'),
    Dropout(0.4),
    Dense(256, activation='relu'),
    Dropout(0.3),
    Dense(len(implant_types), activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Print the model summary
model.summary()
```

```python
# Compute class weights if there's significant class imbalance
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

class_weights = compute_class_weight(
    'balanced',
    classes=np.unique(train_generator.classes),
    y=train_generator.classes
)

class_weight_dict = {i: weight for i, weight in enumerate(class_weights)}
```

```python
# Set up callbacks for better training
callbacks = [
    EarlyStopping(patience=7, restore_best_weights=True),
    ReduceLROnPlateau(factor=0.2, patience=3, min_lr=1e-6),
    ModelCheckpoint('knee_implant_model_best.h5', save_best_only=True)
]

# Training
epochs = 20
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=epochs,
    class_weight=class_weight_dict,  # Add class weights to handle imbalance
    callbacks=callbacks
)

# Save the final model
model.save('knee_implant_model_final.h5')

# Plot the training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Validation'])

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Validation'])
plt.tight_layout()
plt.show()
```

```python
# Evaluate on the test set
test_loss, test_acc = model.evaluate(test_generator)
print(f'Test accuracy: {test_acc:.4f}')

# Calculate and visualize the confusion matrix
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# Get predictions
test_generator.reset()
y_pred = model.predict(test_generator, verbose=1)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = test_generator.classes

# Print classification report
class_names = list(idx_to_class.values())
print(classification_report(y_true, y_pred_classes, target_names=class_names))

# Plot confusion matrix
cm = confusion_matrix(y_true, y_pred_classes)
plt.figure(figsize=(16, 14))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.tight_layout()
plt.savefig('confusion_matrix.png')
plt.show()
```

```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Set path to data
base_path = "/content/drive/MyDrive/Knee"

# 1. DATA INSPECTION - Check which folders actually have images
valid_classes = []
class_sizes = {}

for folder in os.listdir(base_path):
    folder_path = os.path.join(base_path, folder)
    if os.path.isdir(folder_path):
        files = [f for f in os.listdir(folder_path)
                if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp'))]

        # Only include folders that have images
        if len(files) > 0:
            valid_classes.append(folder)
            class_sizes[folder] = len(files)
            print(f"{folder}: {len(files)} images")
        else:
            print(f"⚠️ {folder}: NO IMAGES - will be excluded")

# 2. CREATE FIXED IMAGE GENERATORS - With simpler augmentation
img_size = 224
batch_size = 16

# Simple augmentation focused on preserving implant features
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=5,        # Minimal rotation
    width_shift_range=0.05,  # Small shifts
    height_shift_range=0.05,
    zoom_range=0.05,         # Minimal zoom
    brightness_range=[0.95, 1.05],
    fill_mode='nearest'
)

# Create generators using only valid classes
train_generator = train_datagen.flow_from_directory(
    base_path,
    target_size=(img_size, img_size),
    batch_size=batch_size,
    classes=valid_classes,  # Only use folders with images
    class_mode='categorical',
    subset='training',
    shuffle=True
)

validation_generator = train_datagen.flow_from_directory(
    base_path,
    target_size=(img_size, img_size),
    batch_size=batch_size,
    classes=valid_classes,  # Only use folders with images
    class_mode='categorical',
    subset='validation',
    shuffle=False
)

# 3. SAFETY CHECK - Visualize sample batch to verify data
def show_batch(generator, title):
    x_batch, y_batch = next(iter(generator))
    plt.figure(figsize=(10, 6))

    for i in range(min(9, len(x_batch))):
        plt.subplot(3, 3, i+1)
        plt.imshow(x_batch[i])
        label_idx = np.argmax(y_batch[i])
        class_name = list(generator.class_indices.keys())[list(generator.class_indices.values()).index(label_idx)]
        plt.title(f"{class_name[:10]}...")
        plt.axis('off')

    plt.suptitle(title)
    plt.tight_layout()
    plt.show()

show_batch(train_generator, "Training Samples")

# 4. SAFE CLASS WEIGHTS - Handle empty folders and prevent division by zero
def safe_class_weights(class_sizes, class_indices):
    total_samples = sum(class_sizes.values())
    weights = {}

    for class_name, index in class_indices.items():
        if class_name in class_sizes and class_sizes[class_name] > 0:
            # Apply smoothing factor to prevent extreme weights
            weight = (total_samples / (len(class_sizes) * class_sizes[class_name]))
            # Cap weight to prevent numerical issues
            weights[index] = min(weight, 10.0)
        else:
            weights[index] = 1.0  # Default weight if missing

    return weights

class_weights = safe_class_weights(class_sizes, train_generator.class_indices)

# 5. CREATE STABLE MODEL - Use a simpler architecture with better defaults
# Start with EfficientNetB0 (smaller than B3) - faster and less prone to overfitting
base_model = EfficientNetB0(
    weights='imagenet',
    include_top=False,
    input_shape=(img_size, img_size, 3)
)

# Freeze base model
base_model.trainable = False

# Add classification head with proper regularization
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)  # Strong dropout to prevent overfitting
predictions = Dense(len(valid_classes), activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=predictions)

# Compile with numerically stable settings
model.compile(
    optimizer=Adam(learning_rate=0.0001),  # Lower learning rate
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# 6. TRAIN WITH SAFEGUARDS - Careful monitoring to catch problems
callbacks = [
    # Stop if NaN loss occurs
    tf.keras.callbacks.TerminateOnNaN(),

    # Reduce learning rate when plateauing
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.2,
        patience=3,
        min_lr=1e-6,
        verbose=1
    ),

    # Early stopping with patience
    EarlyStopping(
        monitor='val_accuracy',
        patience=7,
        restore_best_weights=True,
        verbose=1
    )
]

# Training with careful monitoring
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=20,
    class_weight=class_weights,
    callbacks=callbacks,
    verbose=1
)

# 7. EVALUATE AND VISUALIZE RESULTS
# Plot training history
def plot_history(history):
    plt.figure(figsize=(12, 4))

    # Plot accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'])

    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'])

    plt.tight_layout()
    plt.show()

plot_history(history)

# 8. CONFUSION MATRIX - See which implants are confused with each other
def plot_confusion_matrix():
    # Get true labels
    validation_generator.reset()
    y_true = validation_generator.classes

    # Get predicted labels
    y_pred = model.predict(validation_generator)
    y_pred_classes = np.argmax(y_pred, axis=1)

    # Create confusion matrix
    from sklearn.metrics import confusion_matrix
    import seaborn as sns

    cm = confusion_matrix(y_true, y_pred_classes)

    # Normalize confusion matrix
    cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    # Create labels
    class_names = list(validation_generator.class_indices.keys())
    short_names = [name[:10] + "..." if len(name) > 10 else name for name in class_names]

    # Plot
    plt.figure(figsize=(12, 10))
    sns.heatmap(
        cm_norm,
        annot=True,
        fmt='.2f',
        cmap='Blues',
        xticklabels=short_names,
        yticklabels=short_names
    )
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Normalized Confusion Matrix')
    plt.tight_layout()
    plt.show()

plot_confusion_matrix()

# 9. PREDICTION FUNCTION - For using the trained model
def predict_implant(img_path):
    from tensorflow.keras.preprocessing import image

    # Load and preprocess image
    img = image.load_img(img_path, target_size=(img_size, img_size))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = img_array / 255.0

    # Get predictions
    preds = model.predict(img_array)[0]

    # Get top 3 predictions
    top_indices = np.argsort(preds)[-3:][::-1]
    top_preds = [(list(train_generator.class_indices.keys())[i], float(preds[i]) * 100)
                for i in top_indices]

    # Display prediction with image
    plt.figure(figsize=(6, 8))
    plt.imshow(img)
    plt.title(f"Top prediction: {top_preds[0][0]}\n{top_preds[0][1]:.1f}%")
    plt.axis('off')

    # Show other predictions as text
    info = "\n".join([f"{name}: {conf:.1f}%" for name, conf in top_preds[1:]])
    plt.xlabel(f"Other possibilities:\n{info}")
    plt.tight_layout()
    plt.show()

    return top_preds

# Save the model properly
model.save('knee_implant_classifier.keras')  # Use .keras extension as recommended
```
