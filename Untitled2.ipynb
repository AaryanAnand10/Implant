```python
# Import libraries
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, BatchNormalization, Dropout
from tensorflow.keras.layers import Flatten, Dense, GlobalAveragePooling2D, Concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Make sure TensorFlow is working properly
print(f"TensorFlow version: {tf.__version__}")
print(f"GPU available: {len(tf.config.list_physical_devices('GPU')) > 0}")

# Set memory growth to avoid OOM issues
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print("Memory growth set to True for GPUs")
    except RuntimeError as e:
        print(e)

# Path to the knee implant folder
base_path = "/content/drive/MyDrive/Knee"
print("Step 1: Basic setup complete")
```

```python
# Create list of valid folders with image counts
valid_classes = []
class_counts = {}

for folder in os.listdir(base_path):
    folder_path = os.path.join(base_path, folder)
    if os.path.isdir(folder_path):
        images = [f for f in os.listdir(folder_path)
                 if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp'))]
        if len(images) > 0:
            valid_classes.append(folder)
            class_counts[folder] = len(images)
            print(f"{folder}: {len(images)} images")

num_classes = len(valid_classes)
print(f"\nTotal classes: {num_classes}")

# Visualize class distribution
plt.figure(figsize=(15, 6))
plt.bar(class_counts.keys(), class_counts.values())
plt.xticks(rotation=90)
plt.title("Number of Images per Implant Type")
plt.tight_layout()
plt.savefig('class_distribution.png')
plt.show()

print("Step 2: Data analysis complete")
```

```python
# Image dimensions and batch size
img_size = 224
batch_size = 16

# ADVANCED: Two-stage preprocessing for medical images
def enhance_medical_image(img):
    """Custom preprocessing for medical implant images to enhance features"""
    # Convert to float
    img = tf.cast(img, tf.float32)

    # Normalize to 0-1
    img_norm = img / 255.0

    # Local contrast enhancement
    mean = tf.reduce_mean(img_norm)
    std = tf.math.reduce_std(img_norm) + 1e-8
    img_norm = (img_norm - mean) / std * 0.2 + 0.5

    # Clip values
    img_norm = tf.clip_by_value(img_norm, 0.0, 1.0)

    return img_norm

# ADVANCED: Specialized medical image augmentation
train_datagen = ImageDataGenerator(
    preprocessing_function=enhance_medical_image,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    brightness_range=[0.9, 1.1],
    horizontal_flip=True,
    # ADVANCED: Targeted augmentation for implants
    shear_range=0.05,        # Small shear to preserve implant geometry
    fill_mode='constant',    # Black padding for medical images
    cval=0,                  # Black padding value
    validation_split=0.2
)

validation_datagen = ImageDataGenerator(
    preprocessing_function=enhance_medical_image,
    validation_split=0.2
)

# Create train generator with shuffle and class balancing
train_generator = train_datagen.flow_from_directory(
    base_path,
    target_size=(img_size, img_size),
    batch_size=batch_size,
    classes=valid_classes,
    class_mode='categorical',
    subset='training',
    shuffle=True
)

validation_generator = validation_datagen.flow_from_directory(
    base_path,
    target_size=(img_size, img_size),
    batch_size=batch_size,
    classes=valid_classes,
    class_mode='categorical',
    subset='validation',
    shuffle=False
)

# Display a batch to verify data loading
def show_batch(generator, title):
    x_batch, y_batch = next(iter(generator))
    plt.figure(figsize=(10, 6))

    for i in range(min(9, len(x_batch))):
        plt.subplot(3, 3, i+1)
        plt.imshow(x_batch[i])
        label_idx = np.argmax(y_batch[i])
        class_name = list(generator.class_indices.keys())[list(generator.class_indices.values()).index(label_idx)]
        plt.title(f"{class_name[:10]}...")
        plt.axis('off')

    plt.suptitle(title)
    plt.tight_layout()
    plt.show()

show_batch(train_generator, "Training Samples with Enhancement")
print("Step 3: Data generators created")
```

```python
# ADVANCED: Adaptive class weights with smoothing
def create_adaptive_class_weights(class_counts):
    """Creates balanced class weights with adaptive smoothing"""
    total_samples = sum(class_counts.values())
    weights = {}

    # Find corresponding indices for classes
    class_indices = train_generator.class_indices

    for class_name, count in class_counts.items():
        if count > 0:
            # Get class index
            idx = class_indices[class_name]

            # Calculate raw inverse frequency
            raw_weight = total_samples / (len(class_counts) * count)

            # Apply adaptive smoothing - more smoothing for extreme values
            if raw_weight > 3:
                # Heavy smoothing for very rare classes
                weight = 3 + (raw_weight - 3) * 0.3
            elif raw_weight < 0.5:
                # Gentle boost for common classes
                weight = 0.5 + (raw_weight - 0.5) * 0.7
            else:
                # Moderate smoothing for average classes
                weight = raw_weight

            # Cap weight to prevent numerical issues
            weights[idx] = min(weight, 5.0)
        else:
            weights[idx] = 1.0

    return weights

class_weights = create_adaptive_class_weights(class_counts)

# Print class weights
print("Class weights:")
for class_name, idx in train_generator.class_indices.items():
    print(f"{class_name}: {class_weights.get(idx, 1.0):.2f}")

print("Step 4: Class weights calculated")
```

```python
# ADVANCED: Multi-path CNN architecture combining custom CNN and efficient networks
def create_advanced_cnn():
    # Input layer
    input_img = Input(shape=(img_size, img_size, 3))

    # Path 1: Custom CNN layers with regularization
    # Initial convolution with smaller filters
    x1 = Conv2D(32, (3, 3), padding='same', activation='relu')(input_img)
    x1 = BatchNormalization()(x1)
    x1 = Conv2D(32, (3, 3), padding='same', activation='relu')(x1)
    x1 = BatchNormalization()(x1)
    x1 = MaxPooling2D(pool_size=(2, 2))(x1)
    x1 = Dropout(0.25)(x1)

    # Second convolution block
    x1 = Conv2D(64, (3, 3), padding='same', activation='relu')(x1)
    x1 = BatchNormalization()(x1)
    x1 = Conv2D(64, (3, 3), padding='same', activation='relu')(x1)
    x1 = BatchNormalization()(x1)
    x1 = MaxPooling2D(pool_size=(2, 2))(x1)
    x1 = Dropout(0.25)(x1)

    # Third convolution block
    x1 = Conv2D(128, (3, 3), padding='same', activation='relu')(x1)
    x1 = BatchNormalization()(x1)
    x1 = MaxPooling2D(pool_size=(2, 2))(x1)
    x1 = Dropout(0.25)(x1)

    # Flatten CNN output
    x1 = Flatten()(x1)

    # Path 2: Pre-trained model for feature extraction (EfficientNet)
    base_efficient = EfficientNetB0(
        weights='imagenet',
        include_top=False,
        input_tensor=input_img
    )
    base_efficient.trainable = False  # Freeze initially
    x2 = base_efficient.output
    x2 = GlobalAveragePooling2D()(x2)

    # ADVANCED: Feature fusion - combine both paths
    combined = Concatenate()([x1, x2])

    # ADVANCED: Multi-stage dense layers with graduated dropout
    x = Dense(512, activation='relu')(combined)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    x = Dense(256, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.4)(x)
    x = Dense(128, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)

    # Output layer
    output = Dense(num_classes, activation='softmax')(x)

    # Create model
    model = Model(inputs=input_img, outputs=output)

    return model, base_efficient

# Create model
model, base_efficient = create_advanced_cnn()

# ADVANCED: Compile with gradient clipping for stability
optimizer = Adam(learning_rate=0.001, clipnorm=1.0)
model.compile(
    optimizer=optimizer,
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Print model summary
model.summary()
print("Step 5: Model architecture created")
```

```python
# ADVANCED: Comprehensive callbacks system
callbacks = [
    # Early stopping with adequate patience
    EarlyStopping(
        monitor='val_accuracy',
        patience=15,
        restore_best_weights=True,
        verbose=1
    ),

    # ADVANCED: Cyclic learning rate reduction
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,   # Halve the learning rate
        patience=5,
        min_lr=1e-7,
        verbose=1
    ),

    # Checkpoint best model
    ModelCheckpoint(
        'knee_implant_best.keras',
        monitor='val_accuracy',
        save_best_only=True,
        verbose=1
    ),

    # ADVANCED: Terminate on NaN to catch numerical instability
    tf.keras.callbacks.TerminateOnNaN()
]

print("Step 6: Training callbacks configured")
```

```python
print("Starting Phase 1: Training with frozen base model")

# First phase: Train with frozen base model
history1 = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=10,  # Reduced for testing, increase to 20 for full training
    class_weight=class_weights,
    callbacks=callbacks,
    verbose=1
)

# Plot initial training results
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history1.history['accuracy'])
plt.plot(history1.history['val_accuracy'])
plt.title('Phase 1 Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'])

plt.subplot(1, 2, 2)
plt.plot(history1.history['loss'])
plt.plot(history1.history['val_loss'])
plt.title('Phase 1 Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'])
plt.tight_layout()
plt.show()

print("Step 7: Phase 1 training complete")
```

```python
print("Starting Phase 2: Fine-tuning with partial unfreezing")

# Unfreeze the base model for fine-tuning
base_efficient.trainable = True

# Freeze early layers (train only the last 30%)
for layer in base_efficient.layers[:int(len(base_efficient.layers) * 0.7)]:
    layer.trainable = False

# Recompile with lower learning rate for fine-tuning
optimizer = Adam(learning_rate=0.0001, clipnorm=1.0)
model.compile(
    optimizer=optimizer,
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Second phase: Fine-tune with partial unfreezing
history2 = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=10,  # Reduced for testing, increase to 30 for full training
    initial_epoch=len(history1.history['accuracy']),
    class_weight=class_weights,
    callbacks=callbacks,
    verbose=1
)

# Save the final model
model.save('knee_implant_final.keras')

print("Step 8: Phase 2 training complete, model saved")
```

```python
# Combine training histories
history = {}
history['accuracy'] = history1.history['accuracy'] + history2.history['accuracy']
history['val_accuracy'] = history1.history['val_accuracy'] + history2.history['val_accuracy']
history['loss'] = history1.history['loss'] + history2.history['loss']
history['val_loss'] = history1.history['val_loss'] + history2.history['val_loss']

# Plot complete training history with epochs marker for fine-tuning phase
plt.figure(figsize=(14, 5))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(history['accuracy'])
plt.plot(history['val_accuracy'])
plt.axvline(x=len(history1.history['accuracy'])-1, color='r', linestyle='--')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation', 'Start Fine-tuning'])

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history['loss'])
plt.plot(history['val_loss'])
plt.axvline(x=len(history1.history['loss'])-1, color='r', linestyle='--')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation', 'Start Fine-tuning'])

plt.tight_layout()
plt.savefig('knee_implant_training.png')
plt.show()

# Evaluate model on validation set
print("\nFinal Model Evaluation:")
validation_generator.reset()
loss, accuracy = model.evaluate(validation_generator)
print(f"Validation accuracy: {accuracy:.4f}")

# Generate confusion matrix and classification report
validation_generator.reset()
y_pred = model.predict(validation_generator)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = validation_generator.classes

# Print classification report
class_names = list(validation_generator.class_indices.keys())
print("\nClassification Report:")
print(classification_report(y_true, y_pred_classes, target_names=class_names))

# Plot confusion matrix
cm = confusion_matrix(y_true, y_pred_classes)
plt.figure(figsize=(15, 12))

# Use shorter names for readability
short_names = [name[:12] + '...' if len(name) > 12 else name for name in class_names]

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=short_names,
    yticklabels=short_names
)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('confusion_matrix.png')
plt.show()

print("Step 9: Model evaluation complete")
```

```python
# Advanced prediction function with test-time augmentation
def predict_with_tta(image_path, num_augmentations=10):
    """
    Test-Time Augmentation for more reliable predictions
    """
    from tensorflow.keras.preprocessing import image
    import numpy as np

    # Load and preprocess image
    img = image.load_img(image_path, target_size=(img_size, img_size))
    img_array = image.img_to_array(img)

    # Apply core preprocessing
    original_img = enhance_medical_image(img_array)
    original_img = np.expand_dims(original_img, axis=0)

    # Create augmentation generator for test images
    test_datagen = ImageDataGenerator(
        rotation_range=10,
        width_shift_range=0.05,
        height_shift_range=0.05,
        zoom_range=0.05,
        horizontal_flip=True,
        fill_mode='constant',
        cval=0
    )

    # Generate augmented versions
    aug_images = [original_img[0]]
    aug_batch = original_img.copy()

    i = 0
    for batch in test_datagen.flow(aug_batch, batch_size=1):
        aug_img = batch[0]
        aug_images.append(aug_img)
        i += 1
        if i >= num_augmentations-1:
            break

    # Convert to array
    aug_images = np.array(aug_images)

    # Get predictions for all augmented versions
    predictions = model.predict(aug_images)

    # Average the predictions
    avg_pred = np.mean(predictions, axis=0)

    # Get top predictions
    top_indices = np.argsort(avg_pred)[-3:][::-1]
    class_names = list(validation_generator.class_indices.keys())
    top_preds = [(class_names[i], float(avg_pred[i]) * 100) for i in top_indices]

    # Display results
    plt.figure(figsize=(12, 6))

    # Show original image with prediction
    plt.subplot(1, 2, 1)
    plt.imshow(img)
    plt.title(f"Prediction: {top_preds[0][0]}\nConfidence: {top_preds[0][1]:.1f}%")
    plt.axis('off')

    # Show augmentations
    plt.subplot(1, 2, 2)
    grid = np.hstack([aug_images[1], aug_images[2], aug_images[3]])
    plt.imshow(grid)
    plt.title("Augmented Versions (Sample)")
    plt.axis('off')

    plt.tight_layout()
    plt.show()

    return top_preds

print("Step 10: Prediction function with test-time augmentation ready")
print("You can now use predict_with_tta('/path/to/image.jpg') to classify new images")
```
